```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from functions import *
```

```python
datastations = ['USW00022536','USW00023188','USW00014922','RSM00030710'] 
```

```python
minn_tmax = get_obs('USW00014922.dly', 'TMAX')
minn_tmin = get_obs('USW00014922.dly', 'TMIN')

sandi_tmax = get_obs('USW00023188.dly', 'TMAX')
sandi_tmin = get_obs('USW00023188.dly', 'TMIN')
```

```python
fill_nans(minn_tmax)
fill_nans(minn_tmin)
fill_nans(sandi_tmax)
fill_nans(sandi_tmin);
```

```python
years = np.arange(1940, 2014+1)
```

```python
minn_tmax_all = np.vstack([selectyear(minn_tmax, year)[:365] for year in years])
minn_tmin_all = np.vstack([selectyear(minn_tmin, year)[:365] for year in years])

sandi_tmax_all = np.vstack([selectyear(sandi_tmax, year)[:365] for year in years])
sandi_tmin_all = np.vstack([selectyear(sandi_tmin, year)[:365] for year in years])
```

```python
minn_mean = np.mean(minn_tmax_all, axis=1)
minn_warmest = years[np.argmax(minn_mean)]
minn_warmest
```

```python
sandi_mean = np.mean(sandi_tmin_all, axis=1)
sandi_coldest = years[np.argmin(sandi_mean)]
sandi_coldest
```

```python
days = np.arange(1, 366+1)

plt.figure(figsize=(12,6))

plt.fill_between(days,
                selectyear(minn_tmax,minn_warmest),
                selectyear(minn_tmin,minn_warmest),
                color='b', alpha = 0.4)

plt.fill_between(days,
                selectyear(sandi_tmax,sandi_coldest),
                selectyear(sandi_tmin,sandi_coldest),
                 color='r', alpha = 0.4)

plt.axis(xmax=366)
plt.title('{} in Minneapolis vs. {} in Sandiego'.format(minn_warmest, sandi_coldest))
```
