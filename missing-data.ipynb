```python
def fill_nans(data):
    dates_float = data['date'].astype(np.float64)
    
    nan = np.isnan(data['value'])
    data['value'][nan] = np.interp(dates_float[nan], dates_float[~nan], data['value'][~nan])
    
    return data
```

```python
data = lihue_tmax
data_new = fill_nans(data)
```
